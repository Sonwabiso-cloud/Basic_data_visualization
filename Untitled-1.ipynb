```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
df= pd.read_csv('cleaned_stock_prices.csv')
df['date']= pd.to_datetime(df['date'])
#overview of dataset
df.head()
#df.info()
#df.describe()

#bar plot of stock prices by sector
top_stocks = df.groupby("symbol")["close"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_stocks.plot(kind="bar")
plt.title("Top 10 Stocks by Average Closing Price")
plt.xlabel("Stock Symbol")
plt.ylabel("Average Closing Price")
plt.show()

#line plot of stock prices over time for a specific stock
plt.figure(figsize=(10, 5))
specific_stock = df[df['symbol'] == 'AAPL']              
plt.plot(specific_stock['date'], specific_stock['close'], label='AAPL')
plt.title("AAPL Stock Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.show()

#line chart of average closing price over time
daily_avg = df.groupby("date")["close"].mean()
plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg.values)
plt.title("Average Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Average Closing Price")
plt.show()

#scatter plot of opening vs closing prices 
plt.figure(figsize=(10, 5))
sns.scatterplot(x="open", y="close", data=df)
plt.title("Opening Price vs Closing Price")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()

#scatter plot with hue
plt.figure(figsize=(10, 5))
sns.scatterplot(x="open", y="close", hue="volume", data=df, legend=False)
plt.title("Open vs Close Prices Colored by Volume")
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()



#histogram of daily returns
df['daily_return'] = df['close'].pct_change()
plt.figure(figsize=(10, 5))
sns.histplot(df['daily_return'].dropna(), bins=50, kde=True)
plt.title("Histogram of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.show()
```
